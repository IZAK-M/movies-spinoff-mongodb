### Création d'une collection Movies et de documents à partir du CSV Movies small avec Python3

#### Création de la BD Spinoff_3

```python
import pymongo, csv

myclient = pymongo.MongoClient("mongodb://localhost:27017/")

mydb = myclient["Spinoff_3"]
```

-

#### Insertion a partir des CSV

```python
def insert_donnees(csvf, collection):
    data = []
    with open(csvf, newline='', encoding='UTF-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    
        mydb[collection].insert_many(data)
```

```python
insert_donnees("movies-small/movies.csv", "movies")
```

```python
insert_donnees("movies-small/links.csv", "links")
```

```python
insert_donnees("movies-small/ratings.csv", "ratings")
```

```python
insert_donnees("movies-small/tags.csv", "tags")
```
